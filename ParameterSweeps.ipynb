```python
import pandas as pd
import numpy as np
```

```python
# read stored dataframe from the parameter sweeps, dataframe made in basline3.py, data created with cycle_params.py
df = pd.read_pickle("dataframes/baseline3.pkl")
df
```

```python
# create a ranking groped on parameters, with values from start of generation, stop, mean, lowest and difference
# between start and stop
start = pd.Series(df['MSE'].groupby(level=['param', 'value']).first())
mean = pd.Series(df['MSE'].groupby(level=['param', 'value']).mean())
stop = pd.Series(df['MSE'].groupby(level=['param', 'value']).last())
diff = pd.Series(start - stop)
lowest = pd.Series(df.loc[df['MSE'].groupby(level=['param', 'value']).idxmin()]['MSE'])
lowest.index = lowest.index.droplevel(['gen'])
ranked = pd.DataFrame(index=start.index)
ranked['start'] = start
ranked['mean'] = mean
ranked['lowest'] = lowest
ranked['stop'] = stop
ranked['diff'] = diff
ranked.apply(lambda x: x ** 0.5)
```
